```python
#function to search for the book the user gives as input
def book_search(user_input,dataframe):
    cleaned_input = user_input.lower().strip()
    matching_books = dataframe[dataframe['Book'].str.lower().str.contains(cleaned_input)]

    return matching_books

#test outcome of book search function
book_search(' harry potter',df)
```

## Recommendations using TF-IDF

```python
#combine text features into a single column
#weighting genre more results in better recommendations
df['combined'] = (df['Author'] + " " + (df['Genres'].astype(str) + " ")*2 + (df['Description'].astype(str)))
```

```python
#vectorise the features
from sklearn.feature_extraction.text import TfidfVectorizer
vectoriser = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectoriser.fit_transform(df['combined'])
print(tfidf_matrix.shape)
```

```python
#compute the similarities using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
from rapidfuzz import fuzz

#function to determine if two book titles are very similar
def is_too_similar (title1, title2,threshold = 50):
    return fuzz.token_set_ratio(title1.lower(), title2.lower()) >= threshold
```

```python
#genre function to make sure that recommended books share at least one genre
def share_genres(book_genres, candidate_genres):
    set1 = set(g.strip().lower() for g in book_genres.split(','))
    set2 = set(g.strip().lower() for g in candidate_genres.split(','))
    return len(set1.intersection(set2)) > 0
```

```python

#obtain list of similar books to input book
def get_similar_books(user_input):
    found_books = book_search(user_input,df)

    if found_books.empty:
        return None,None

    #pick first match
    idx = found_books.index[0]
    chosen_book = df.loc[idx]
    matrix_idx = df.index.get_loc(idx)

   #get similarity scores for chosen book
    similarity_scores = list(enumerate(similarity_matrix[matrix_idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    #get top similar book titles
    list_similar_books = []

    for i,score in similarity_scores:
        candidate_title = df.iloc[i]['Book']
        #filter out book itself
        if i == idx:
            continue
        #filter out too similar books (e.g. same book with slightly different title/different edition)
        if is_too_similar(candidate_title,chosen_book['Book']):
            continue
        #must share at least one genre
        if not share_genres(chosen_book['Genres'],df.iloc[i]['Genres']):
            continue

        list_similar_books.append(df.iloc[i])

        if len(list_similar_books) == 4:
            break

    similar_books = pd.DataFrame(list_similar_books)
    similar_books = similar_books.drop_duplicates(subset = 'Book')

    return similar_books, chosen_book

#main recommendation function
def recommendation():
    user_input = input("Please enter your book title: ")
    print("You entered: ", user_input)
    similar_books, chosen_book = get_similar_books(user_input)

    if similar_books is None or chosen_book is None:
        print(f"No books found matching '{user_input}'")
        return

    book_title = chosen_book['Book']
    print(f"Because you enjoyed {book_title}, you might like one of the following books:")
    for i in range(min(4,len(similar_books))):
        print(f"- {similar_books.iloc[i]["Book"]} by {similar_books.iloc[i]["Author"]}")
```

```python
recommendation()
```

The recommendations above were generated using a TF-IDF-based approach, which finds books that share words in their combined metadata (title, author, genres, description). This works well for obvious similarities but can miss deeper connections between books.

To address this, I built a separate recommendation engine using SBERT embeddings, which represent each book as a dense vector capturing its overall meaning. This approach allows the SBERT engine to suggest books that are conceptually similar, even if they don’t share many exact words, resulting in more diverse and relevant recommendations.

## Recommendations using SBERT

```python
from sentence_transformers import SentenceTransformer

#load model2
model = SentenceTransformer("all-distilroberta-v1")

#calculate embeddings from combined column
recommendation_embeddings = model.encode(df['combined'].tolist(),       convert_to_tensor=True,show_progress_bar=True)
print(recommendation_embeddings.shape)
```

```python
from sentence_transformers import util

#calculate embedding similarities
similarities = util.cos_sim(recommendation_embeddings,recommendation_embeddings)
```

```python
import torch

#get similar book to user input

def get_sim_books(user_input,top_k = 4):
    found_books = book_search(user_input,df)

    if found_books.empty:
        return None,None

    #pick first match
    idx = found_books.index[0]
    chosen_book = df.loc[idx]
    matrix_idx = df.index.get_loc(idx)

   #get similarity scores for chosen book
    sim_scores = similarities[matrix_idx]

    #sort similarity scores descending
    sorted_sim_scores, sorted_indices = torch.sort(sim_scores, descending=True)

    #remove book itself
    nobook = sorted_indices != matrix_idx
    sorted_indices = sorted_indices[nobook]
    sorted_sim_scores = sorted_sim_scores[nobook]

    #convert to list and filter out too similar titles, keep top_k
    filtered_indices = [i for i in sorted_indices.tolist() if not is_too_similar(df.iloc[i]['Book'], chosen_book['Book'])
                        and share_genres(chosen_book['Genres'],df.iloc[i]['Genres'])]

    #filter the indices further: check if a book is mid-series and replace it, if possible, with the first book of the series
    final_indices = []
    for index in filtered_indices:
        if len(final_indices) == top_k:
            break
        title = df.iloc[index]['Book']
        original_title = find_first_book(title)
        associated_index = int(df.index[df['Book'] == original_title][0])
        if(associated_index not in final_indices):
            final_indices.append(associated_index)


    #get dataset entries corresponding to these indices
    similar_books = df.loc[final_indices]

    return similar_books, chosen_book

#main recommendation function
def recommend():
    user_input = input("Please enter your book title: ")
    print("You entered: ", user_input)
    similar_books, chosen_book = get_sim_books(user_input)

    if similar_books is None or chosen_book is None:
        print(f"No books found matching '{user_input}'")
        return

    book_title = chosen_book['Book']
    print(f"Because you enjoyed {book_title}, you might like one of the following books:")
    for i in range(len(similar_books)):
        print(f"- {similar_books.iloc[i]["Book"]} by {similar_books.iloc[i]["Author"]}")
```

```python
recommend()
```

The recommendations include books that are either from the same universe/author (e.g., The Tales of Beedle the Bard, Fantastic Beasts) or share similar themes and target audience (e.g., Magic Kingdom for Sale/Sold, A Hat Full of Sky). This demonstrates that the recommendation engine considers both direct connections and broader thematic similarities, making the suggestions relevant even for readers who have already explored the main series.

```python
import re

#obtain series number of a book, return none if there is no series number
def get_series_number(book_title):
    series_num = re.search(r'#(\d+)', book_title)
    if series_num:
        return int(series_num.group(1))
    return None

get_series_number('Harry Potter and the Half-Blood Prince (Harry Potter, #6)')
```

```python
#choose first book in series
def find_first_book(book_title):
    series_num = get_series_number(book_title)

    if series_num is None or series_num == 1:
        return book_title

    #extract series title
    series_match = re.search(r'\((.*)#(\d+)\)', book_title)

    if not series_match:
        return book_title

    series_name = str(series_match.group(1))

    #find first book in series in dataset
    first_books = df[df['Book'].str.contains(series_name + '#1')]

    if not first_books.empty:
        return first_books.iloc[0]['Book']

    return book_title

```
